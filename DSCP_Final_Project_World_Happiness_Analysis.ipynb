```python

# DSCP Final Project: World Happiness Analysis
# OBJECTIVE: To analyze if a country's economic strength (GDP)
# directly correlates with its citizens' happiness score.

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# PSEUDOCODE:
# 1. Load data (or create dummy data for demonstration)
# 2. Clean data (check for nulls)
# 3. Algorithm: Categorize GDP into High/Low & Group by Region
# 4. Visualize: Scatter Plot & Bar Chart

data = {
    'Country': ['Finland', 'Denmark', 'Norway', 'USA', 'Germany',
                'Brazil', 'China', 'India', 'Chad', 'Afghanistan'],
    'Region': ['Western Europe', 'Western Europe', 'Western Europe', 'North America', 'Western Europe',
               'Latin America', 'East Asia', 'South Asia', 'Sub-Saharan Africa', 'South Asia'],
    'Happiness Score': [7.8, 7.6, 7.5, 6.9, 7.0, 6.1, 5.1, 4.0, 3.5, 2.5],
    'GDP per Capita': [1.3, 1.35, 1.45, 1.4, 1.3, 0.9, 0.95, 0.7, 0.3, 0.25]
}
df = pd.DataFrame(data)

# Algorithm 1: Data Cleaning

if df.isnull().sum().sum() > 0:
    print("Missing values found. Filling with mean...")
    df.fillna(df.mean(numeric_only=True), inplace=True)

# Algorithm 2: Transformation & Grouping

def categorize_gdp(gdp_value):
    if gdp_value > 1.0:
        return 'High Income'
    else:
        return 'Low/Medium Income'

df['Income_Level'] = df['GDP per Capita'].apply(categorize_gdp)

# Calculate average happiness by Region
regional_happiness = df.groupby('Region')['Happiness Score'].mean().sort_values(ascending=False).reset_index()

print("Top 3 Happiest Regions:")
print(regional_happiness.head(3))


# Scatter Plot (Money vs Happiness)
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='GDP per Capita', y='Happiness Score', hue='Region', s=100)
plt.title('Correlation: GDP vs Happiness Score', fontsize=15)
plt.xlabel('GDP per Capita')
plt.ylabel('Happiness Score')
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
sns.barplot(data=regional_happiness, x='Happiness Score', y='Region', palette='viridis')
plt.title('Average Happiness Score by Region', fontsize=15)
plt.show()
```

```python
correlation = df['GDP per Capita'].corr(df['Happiness Score'])
print(f"The Correlation Coefficient is: {correlation:.2f}")
```
